```python
import os
import numpy as np
import pandas as pd

ImgSplits_path = os.getcwd() + "\Stanford40\ImageSplits"
Imgpath = os.getcwd() + "\Stanford40\JPEGImages"

actions_path = ImgSplits_path + "\\actions.txt"

print(actions_path)
#read and parse the .csv actions.txt file to determine the 40 action names 
df_action = pd.read_csv(actions_path, delimiter = '\t+')

# Enumerate the actions with labels from 0 to 39
df_action['label'] = range(0, len(df_action))
df_action.head()
```

```python
#read the action names and their labels
actions = df_action[df_action.columns[0]]
labels = df_action[df_action.columns[2]]

#join the actions and their labels in a dictionary 
action_dict = dict(zip(np.array(actions), np.array(labels)))
```

```python
print(action_dict)
```

```python
#read the training images filenames only
ImgSplits_path = os.getcwd() + "\Stanford40\ImageSplits"
train_path = ImgSplits_path + "\\train.txt"

#read and parse the train.txt file to determine the 4000 train image filenames 
df_train = pd.read_csv(train_path, header=None)
df_train.head()
train_filenames = df_train[df_train.columns[0]]
#print(train_filenames)
```

```python
#read the test images filenames only
ImgSplits_path = os.getcwd() + "\Stanford40\ImageSplits"
test_path = ImgSplits_path + "\\test.txt"

#read and parse the train.txt file to determine the 5532 test image filenames 
df_test = pd.read_csv(test_path, header=None)
df_test.head()
test_filenames = df_test[df_test.columns[0]]
#print(test_filenames)
```

```python

train_labels = []
#for loop to assign a label to every train filename according to the name of the file 
for train_filename in train_filenames:
    for action in actions:
        if action in train_filename:
            train_labels.append(action_dict[action])

print(train_labels)
```

```python
print(np.array(train_labels).shape)
```

```python
test_labels = []
#for loop to assign a label to every test filename according to the name of the file 
for test_filename in test_filenames:
    for action in actions:
        if action in test_filename:
            test_labels.append(action_dict[action])

#print(test_labels)
```

```python
print(np.array(test_labels).shape)
```

```python
#now read all the train images
import cv2

images_path = os.getcwd() + "\Stanford40\JPEGImages"
train_images = []
for train_filename in train_filenames:
    img_path = os.path.join(images_path, train_filename) 
    img = cv2.imread(img_path)
    img = cv2.resize(img,(128,128))
    
    train_images.append(img)
```

```python
#now read all the test images
import cv2

images_path = os.getcwd() + "\Stanford40\JPEGImages"
test_images = []
for test_filename in test_filenames:
    img_path = os.path.join(images_path, test_filename) 
    img = cv2.imread(img_path)
    img = cv2.resize(img,(128,128))
    
    test_images.append(img)
```

```python
# Normalize the train images.
train_images = np.array(train_images)
train_images = (train_images/255) - 0.5
#print(type(train_images))

```

```python
print(type(train_images))
print(train_images.shape)
```

```python
# Normalize the test images.
test_images = np.array(test_images)
test_images = (test_images/255) - 0.5
#print(type(test_images))
```

```python
print(type(test_images))
print(test_images.shape)
```

```python
# train and validate using Keras to implement a simple Convolutional Neural Network (CNN) in Python.
# code inspired by https://victorzhou.com/blog/keras-cnn-tutorial/ parameters were modified as below
import tensorflow as tf

num_filters = 1
filter_size = 3
pool_size = 2
model = tf.keras.models.Sequential([
  tf.keras.layers.Conv2D(num_filters, filter_size, input_shape=(128, 128, 3)),
  tf.keras.layers.MaxPooling2D(pool_size=pool_size),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(40, activation='softmax'),
])

model.compile(
  'adam',
  loss='categorical_crossentropy',
  metrics=['accuracy'],
)

model.fit(
  train_images,
  tf.keras.utils.to_categorical(train_labels),
  epochs=10,
  validation_data=(test_images, tf.keras.utils.to_categorical(test_labels)),
)
```
